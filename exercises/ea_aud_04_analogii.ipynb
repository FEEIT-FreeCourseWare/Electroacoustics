1. Најди ја резонантната фреквенција $\omega_0$ на Хелмхолцов резонатор со $r = 20 $cm, $l = 5$ cm, $a = 10$ cm.

$|Z_i| = 0 $ за $\omega = \omega_0$

$|Z_i| = |j\omega m_a + \frac{1}{j\omega C_a}|$

$|Z_i| = \omega m_a - \frac{1}{\omega C_a}$

$|Z_i| = \omega_0 m_a - \frac{1}{\omega_0 C_a} = 0$

$\omega_0 = \frac{1}{\sqrt{m_a C_a}}$

$C_a = \frac{V}{\rho c^2}$

$m_a = \frac{\rho l_{eff}}{S}$

```python
import numpy as np

rho = 1.293  # kg/m3
rhoc = 414  # kg/m2s
c = 343  # m/s
r = 0.2  # m
l = 0.05  # m
a = 0.1  # m

V = 4 * np.pi / 3 * r**3
Ca = V / (rhoc * c)  # m5s/kg - SI edinici
l_eff = l + 2*0.85*a
S = np.pi * a**2
ma = rho * l_eff / S  # kg/m4 - SI edinici

omega_0 = 1 / np.sqrt(Ca * ma)  # rad/s
f0 = omega_0 / (2*np.pi)  # Hz
print(f0)
```

```python
c = 343  # m/s
f = 200  # Hz
lamda = c / f
print(lamda)
```

```python
rhoc = 414 
p = 0.2  # Pa
r = 2  # m
Pa = 2*np.pi*r**2 * p**2/rhoc
print(Pa, ' W')
```

```python
Ra = 860  # SI edinici
q = np.sqrt(Pa / Ra)  # m3/s
V = 0.15 * 0.4 * 0.3  # m3
Ca = V / (rhoc * c)  # SI edinici
pk = q * 1 / (2*np.pi*f * Ca)  # Pa
print(pk, ' Pa')
```
