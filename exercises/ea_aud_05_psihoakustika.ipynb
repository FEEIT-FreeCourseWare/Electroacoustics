(3.3) Простопериодичен тон со $f$ од 100 Hz, има ниво од 80 dB. 
- а) Да се одреди субјективната јачина и гласноста на овој звук.
- б) На кое ниво треба да се засили звукот за да биде двојно погласен? 

$ \Gamma = 2 ^ {\frac{\Lambda - 40}{10}}$

Релевантни се црвените криви - ISO стандард 226:2003. 

Од кривите може да прочитаме дека $\Lambda $ изнесува околу 62 phon.

```python
Lamda = 62  # phon
Gamma = 2 ** ((Lamda - 40) / 10)
print(Gamma, ' son')
```

```python
Gamma2 = 2 * Gamma
print(Gamma2, ' son')
```

$ \Lambda = \frac{10}{\log 2} \log \Gamma + 40$

```python
import numpy as np
Lamda2 = 10 * np.log2(Gamma2) + 40
print(Lamda2, ' phon')
```

Од кривите имаме $L_2 = 85 $ dB.

(3.5) Еден сложен звук содржи три компоненти на фреквенции од 100, 500 и 1000 Hz, со ниво од 60, 70 и 80 dB. Да се пресмета:
- а) вкупната гласност на овој звук
- б) вкупното ниво на овој звук
- в) потребно ниво на тон од 2 kHz за да ја има истата гласност со овој звук.


$ \Gamma = 2 ^ {\frac{\Lambda - 40}{10}}$

Читајќи изофонските криви можеме да ја најдеме јачината на секоја од компонентите: 37, 67, 80 phon.

```python
Lamdas = [37, 67, 80]  # phon
Gammas = []
for Lamda in Lamdas:
    Gamma = 2**((Lamda - 40) / 10)  # son
    print(Lamda, ' phon -> ', Gamma, ' son')
    Gammas.append(Gamma)
```

Вкупна гласност е $\Gamma_{tot} = \sum_i \Gamma_i$ акко тие се во различни октави.

```python
Gamma_tot = sum(Gammas)
print(Gamma_tot, ' son')
```

- б) Вкупно ниво се пресметува преку $J_{tot} = \sum_i J_i$

$L = 20 \log \frac{p}{p_0}$

$L = 10 \log \frac{J}{J_0}$, за $J_0 = 10^{-12} $ W/m$^2$

```python
Ls = [60, 70, 80]  # dB
J0 = 1e-12  # W/m2
Js = []
for L in Ls:
    J = J0 * 10**(L / 10)  # W/m2
    print(L, ' dB -> ', J, ' W/m2')
    Js.append(J)
```

```python
Jtot = sum(Js)
print(Jtot, ' W/m2')
import numpy as np
Ltot = 10 * np.log10(Jtot / J0)
print(Ltot, ' dB')
```

- в) $\Lambda_{2kHz} = 10 \log_2 \Gamma_{tot} + 40$

```python
Lamda_2kHz = 10 * np.log2(Gamma_tot) + 40  # son
print(Lamda_2kHz, ' phon')
```

Од изофонските криви го отчитуваме $L_{2kHz} = 85$ dB

(3.6) Еден звук има три компоненти на фреквенции 200, 1000 и 1500 Hz, со ниво од 60, 55 и 50 dB, соодветно. Да се пресемта:
- а) колку dB би покажал мерачот на бучава со линеарна карактеристика (фонометар)?
- б) колку dB би покажал мерачот на бучава со А карактеристика?
- в) колку изнесува субјективната јачина на овој звук?

- а) Вкупно ниво се пресметува преку $J_{tot} = \sum_i J_i$

```python
Ls = [60, 55, 50]  # dB
J0 = 1e-12  # W/m2
Js = []
for L in Ls:
    J = J0 * 10**(L / 10)  # W/m2
    print(L, ' dB -> ', J, ' W/m2')
    Js.append(J)
Jtot = sum(Js)
print(Jtot, ' W/m2')
Ltot = 10 * np.log10(Jtot / J0)
print(Ltot, ' dB')
```

- б) Ќе ја искористиме А карактеристиката за корекција на нивоата.

Од графиконот, корекцијата изнесува: 
- 200 Hz $\rightarrow$ -11 dB $\rightarrow$ $L_k = 60 - 11 = 49$ dB
- 1000 Hz $\rightarrow$ 0 dB $\rightarrow$ $L_k = 55 $ dB
- 1500 Hz $\rightarrow$ +1 dB $\rightarrow$ $L_k = 50 + 1 = 51$ dB

```python
Ls = [49, 55, 51]  # dB
J0 = 1e-12  # W/m2
Js = []
for L in Ls:
    J = J0 * 10**(L / 10)  # W/m2
    print(L, ' dB -> ', J, ' W/m2')
    Js.append(J)
Jtot = sum(Js)
print(Jtot, ' W/m2')
Ltot = 10 * np.log10(Jtot / J0)
print(Ltot, ' dB')
```

- в) субјективна јачина. Не може да се направи едноставен збир бидејќи две компоненти се во иста октава. Во овој случај се прави собирање на нивоата по октави, а потоа се утврдува јачина.

$J_{1,1.5} = J_1 + J_{1.5}$

```python
Ls = [55, 50]  # dB
J0 = 1e-12  # W/m2
Js = []
for L in Ls:
    J = J0 * 10**(L / 10)  # W/m2
    print(L, ' dB -> ', J, ' W/m2')
    Js.append(J)
J_1_15 = sum(Js)
print(J_1_15, ' W/m2')
L_1_15 = 10 * np.log10(J_1_15 / J0)
print(L_1_15, ' dB')
```

Од изофонските криви исчитуваме фони:
- 200 Hz $\rightarrow$ 60 dB $\rightarrow$ $\Lambda = 50 $ phon
- 1000 Hz и 1500 Hz $\rightarrow$ 56.19 dB $\rightarrow$ $\Lambda = 56.2 $ phon

```python
Lamdas = [50, 56.2]  # phon
Gammas = []
for Lamda in Lamdas:
    Gamma = 2**((Lamda - 40) / 10)  # son
    print(Lamda, ' phon -> ', Gamma, ' son')
    Gammas.append(Gamma)
Gamma_tot = sum(Gammas)
print(Gamma_tot, ' son')
```

- $\Lambda = 10 \log_2 \Gamma_{tot} + 40$

```python
Lamda_tot = 10 * np.log2(Gamma_tot) + 40  # son
print(Lamda_tot, ' phon')
```
