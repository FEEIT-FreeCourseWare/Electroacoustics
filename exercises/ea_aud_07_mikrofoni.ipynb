(5.5) Извор со $P_a$ од 0,68 mW и $f$ од 50 Hz, е поставен близу до ѕид со $\alpha$ од 0,36 на растојание $d$ од 2,6 m. Колкав напон ќе даде микрофон поставен на $l =$ 3 m од изворот, на исто растојание од ѕидот, ако карактеристиката на насоченост му е кардиоида, а осетливоста $T_0$ му е 0,5 mV/Pa.

$ T = T_0 \frac{1 + \cos \theta}{2}$

$u = u_s + u_r$

$u_s = T_0 p_s$

$u_r = T_0 \frac{1 + \cos \theta}{2} p_r$

$ p = \sqrt{\frac{(1 - \alpha) P_a \rho c}{4 \pi r^2}}$

```python
import numpy as np

Pa = 0.68e-3  # W
rhoc = 414  # kg /m2 s
l = 3  # m
ps = np.sqrt(Pa * rhoc / (4 * np.pi * l**2))
T0 = 0.5  # mV /Pa
us = T0 * ps  # mV
print('u_s = ', us, ' mV')
```

```python
alpha = 0.36
d = 2.6  # m
r = np.sqrt(l**2 + (2*d)**2)  # m
print('r =', r, ' m')
pr = np.sqrt((1 - alpha) * Pa * rhoc / (4 * np.pi * r**2))  # Pa
theta = np.arcsin(2*d / r)
print('theta = ', theta / np.pi * 180)
ur = T0 * (1 + np.cos(theta)) / 2 * pr  # mV
print('u_r = ', ur, ' mV')
```

$u = \sqrt{u_s^2 + u_r^2 + 2u_s u_r \cos \varphi}$

```python
c = 343  # m/s
f = 50  # Hz
lamda = c / f  # m
k = 2*np.pi / lamda  # rad/m
d_r = r - l  # m
phi = k * d_r  # rad
print('lamda = ', lamda, ' m')
print('phi = ', phi/np.pi *180, ' deg')
```

```python
u = np.sqrt(us**2 + ur**2 + 2*us*ur*np.cos(phi))
print('u = ', u, ' mV')
```

(5.12) Микрофон со внатрешна отпорност $R_m$ од 60 $\Omega$, се приклучува на засилувач со влезна отпорност $R_i = 500 \Omega$. Да се пресмета за колку dB се разликува напонот на влезот од засилувачот ако микрофонот е приклучен директно и ако се примени трансформатор за прилагодување.

a) $u_i = \frac{R_i}{R_m + R_i} u_m$


```python
Rm = 50
Ri = 500
r1 = Ri / (Rm + Ri)  # ui / um
print(r1)
```

$n = \sqrt{\frac{R_m}{R_i}}$

```python
n = np.sqrt(Rm / Ri)
print(n)
```

$u_p = \frac{u_m}{2}$

$u_i = u_s = \frac{u_p}{n}$

```python
r2 = 1 / (2 * n)  # ui / um
print(r2)
```

```python
print(20 * np.log10(r2 / r1), ' dB')
```
