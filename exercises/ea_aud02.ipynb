1.23 Звучниот извор $S$ е поставен близу до бескрајно крута плоча. $P_a$ e 1,6 mW, $f$ e 400 Hz. Колкав е притисокот во мерната точка $M$?

$p_M = p_S + p_r$

1. цртање на еквивалентен систем без препрека

2. пресметка на амплитудите на брановите кои доаѓаат до М

$ p = \sqrt{\frac{(1 - \alpha) P_a \rho c}{\Omega_z r^2}}$

$\alpha$ - **коефициент на апсорпција**

```python
import numpy as np
Pa = 1.6e-3  # W
l = 3  # m
rhoc = 414  # kg/m2s
p_s = np.sqrt(Pa * rhoc / (4 * np.pi * l**2))
print(Pa, 'W')
print(p_s, 'Pa')
```

```python
d = 0.142  # m
r = 2*d + l
p_r = np.sqrt(Pa * rhoc / (4 * np.pi * r**2)) 
print(p_r, 'Pa')
```

3. чекор - пресметка на фазната разлика

$\Delta \varphi = k \Delta r = \frac{2\pi}{\lambda} \Delta r$

```python
c = 343  # m/s
f = 400  # Hz
lamda = c / f  # m
print(lamda, 'm')
k = 2*np.pi / lamda  # m-1
d_r = l - r
d_phi = k * d_r
print(d_phi, 'rad')
print(d_phi * 180 / np.pi, 'deg')
```

4. фазорско собирање на звучните бранови

$p_M^2 = p_S^2 + p_r^2 - 2p_S p_r \cos\beta$

$p_M^2 = p_S^2 + p_r^2 + 2p_S p_r \cos\varphi$

```python
p_m = np.sqrt(p_s**2 + p_r**2 + 2*p_s*p_r*np.cos(d_phi))  # Pa
print(p_m, 'Pa')
```

**Колоквиумска задача 1.** Звучен извор поставен помеѓу две бесконечно големи плочи емитува звук на $f$ од 500 Hz, со акустичка моќност $P_a$ од 1 mW. Да се најде притисокот во мерната точка $M$.

1. чекор - еквивалентирање на системот без прегради

$ p_M = p_S + p_{S_1} + p_{S_2} $

2. пресметка на амплитудите на брановите кои доаѓаат до М

$ p = \sqrt{\frac{(1 - \alpha) P_a \rho c}{\Omega_z r^2}}$

$\alpha$ - **коефициент на апсорпција**

```python
import numpy as np
Pa = 1e-3  # W
rho_c = 414  # kg/m2s
alpha1 = 0.5
alpha2 = 0.3
omega_z = 4 * np.pi  # rad
c = 343  # m/s
f = 500  # Hz
lamda = c / f  # m
print(lamda, 'm')
l = 3/2 * lamda
d = lamda/2
r1 = np.sqrt((2*d)**2 + l**2)
r2 = r1
print(r1, 'm')
p_s = np.sqrt(Pa * rho_c / (omega_z * l**2))
p_s1 = np.sqrt((1 - alpha1) * Pa * rho_c / (omega_z * r1**2))
p_s2 = np.sqrt((1 - alpha2) * Pa * rho_c / (omega_z * r2**2))
print(p_s, 'Pa')
print(p_s1, 'Pa')
print(p_s2, 'Pa')
```

3. чекор - пресметка на фазната разлика

$\Delta \varphi = k \Delta r = \frac{2\pi}{\lambda} \Delta r$

- брановите од $S_1$ и $S_2$ се во фаза, но не и $S$

```python
p_s12 = p_s1 + p_s2  # брановите се во фаза
d_r = l - r1
k = 2*np.pi / lamda  # rad/m
d_phi_s_s12 = k * d_r  # rad
print(d_phi_s_s12, 'rad')
print(d_phi_s_s12 * 180/np.pi, 'deg')
```

4. фазорско собирање на звучните бранови

$p_M^2 = p_S^2 + p_r^2 - 2p_S p_r \cos\beta$

$p_M^2 = p_S^2 + p_r^2 + 2p_S p_r \cos\varphi$

```python
p_m = np.sqrt(p_s**2 + p_s12**2 + 2*p_s*p_s12*np.cos(d_phi_s_s12))  # Pa
print(p_m, 'Pa')
```
